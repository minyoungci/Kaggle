```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv("train_V2.csv")
train
```

```python
train.info()
```

```python
train.nunique()
```

```python
print("The average person kills {:.4f} players, 99% of people have{} kills or less, while the most kills ever recorded is {}.".format(train['kills'].mean(),
                                                                                                                                      train['kills'].quantile(0.99)
                                                                                                                                      , train['kills'].max()))
```

```python
data = train.copy()
data.loc[data['kills'] > data['kills'].quantile(0.99)] ='8+'
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data['kills'].astype('str').sort_values())
plt.title('kills count')
```

```python
data = train.copy()
data = data[data['kills']==0]

plt.figure(figsize=(10,10))

plt.title("Damage Dealt by 0 killers", fontsize=15)
sns.distplot(data['damageDealt'])
plt.show()
```

```python
print(f"{len(data[data['winPlacePerc']==1])} players {100*len(data[data['winPlacePerc']==1])/ len(train)}")
```

```python
sns.jointplot(x='winPlacePerc', y='kills', data=train,
             height=10,
             ratio=3,
             color='r')
```

```python
kills = train.copy()

kills['killsCategories'] = pd.cut(kills['kills'], [-1, 0, 2, 5, 10, 60], labels=['0_kills', '1-2_kills', '3-5_kills',
                                                                                '6-10_kills',
                                                                                '10+_kills'])

plt.figure(figsize=(10,8))
sns.boxplot(x='killsCategories', y= 'winPlacePerc', data=kills)
plt.show()
```

```python
print(f"The average person walks for {train['walkDistance'].mean()}m, 99% of people have walked {train['walkDistance'].quantile(0.99)}m or less, while the marathoner champion walked for {train['walkDistance'].max()}m")
```

```python
data = train.copy()

data = data[data['walkDistance'] < train['walkDistance'].quantile(0.99)]

plt.figure(figsize=(15,8))
plt.title("Walking Distance Distribution", fontsize=10)

sns.distplot(data['walkDistance'])
plt.show()
```
